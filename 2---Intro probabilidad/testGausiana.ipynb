```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
folder = "./"
#Train data
trainImages = np.load(folder+'train_images.npy')
trainLabels = np.loadtxt(folder+'train_labels.csv', delimiter=',', skiprows=1,dtype=int,usecols=(0,))
#test data
testImages = np.load(folder+'test_images.npy')
testLabels = np.loadtxt(folder+'test_labels.csv', delimiter=',', skiprows=1,dtype=int,usecols=(0,))
```

```python
gauss_params = []
for cat in range(10):
    means= np.mean(trainImages[trainLabels==cat],axis=0)
    stds = np.std(trainImages[trainLabels==cat],axis=0)
    gauss_params.append((means,stds))
x,y = 14,14
```

```python
cat = 0
gauss_params[cat][0][14,14] #0 indica media
```

```python
xs = np.linspace(0,255,255)

for cat in range(10):
    media = gauss_params[cat][0][x,y]
    desvio = gauss_params[cat][1][x,y]
    print(media,desvio)
    pdf = norm(media,desvio)
    plt.plot(xs,pdf.pdf(xs))
plt.show()
```
