```python
#Guardar una clase en disco
import pickle
```

```python
with open("arr6_pick.pkl","wb")as f:#retorna un objeto en f, creando un contexto en python wb->write binary 
    pickle.dump(arr6,f)#guarda en disco dicho objeto
```

```python
with open("arr6_pick.pkl","rb")as f:#retorna un objeto en f, creando un contexto en python rb->read binary 
    arr8= pickle.load(f)#guarda en disco dicho objeto
    print(arr8)
```
